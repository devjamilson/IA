```python
import pandas as pd

data = pd.read_csv('./Datas/Student Mental health.csv')
df_especific = data[['Age', 'What is your course?', 'Choose your gender', 'Do you have Depression?', 'Do you have Anxiety?', 'Do you have Panic attack?']]
df_especific = df_especific.rename(columns={'What is your course?':'Course','Choose your gender':'Gender', 'Do you have Depression?': 'Depression', 'Do you have Anxiety?':'Anxiety',  'Do you have Panic attack?': 'Panic_Attack' })

df_especific
```

```python
#quais cursos foram analisados
cursos_analisados = df_especific['Course'].unique()
cursos_analisados
```

```python
#qual o genero mais presente no curso

# Agrupar por curso e gênero e contar as ocorrências
contagem_genero_por_curso = df_especific.groupby(['Course', 'Gender']).size().reset_index(name='Count')

# Encontrar o gênero mais presente em cada curso
indice_maximo = contagem_genero_por_curso.groupby('Course')['Count'].idxmax()
genero_mais_presente = contagem_genero_por_curso.loc[indice_maximo]

# Exibir o gênero mais presente em cada curso
print("Gênero mais presente em cada curso:")
print(genero_mais_presente)
```
